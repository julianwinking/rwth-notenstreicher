```python
def berechne_schnitt(data):
    return (data["Note"] * data["Credits"]).sum() / data["Credits"].sum()

# Aktueller Schnitt
aktueller_schnitt = berechne_schnitt(df)
print(f"Aktueller Schnitt: {aktueller_schnitt:.3f}")

# Schnitt pro Semester
def berechne_schnitt_pro_semester(data):
    return data.groupby("Semester").apply(berechne_schnitt)
schnitt_pro_semester = berechne_schnitt_pro_semester(df)
print("\nSchnitt pro Semester:")
print(schnitt_pro_semester)
```

```python
beste_verbesserung = 0
zu_streichendes_fach = None
neuer_bester_schnitt = aktueller_schnitt

for i in range(len(df)):
    df_temp = df.drop(index=i)
    neuer_schnitt = berechne_schnitt(df_temp)
    verbesserung = aktueller_schnitt - neuer_schnitt
    if verbesserung > beste_verbesserung:
        beste_verbesserung = verbesserung
        zu_streichendes_fach = df.iloc[i]["Fach"]
        neuer_bester_schnitt = neuer_schnitt

if zu_streichendes_fach:
    print(f"Wenn du '{zu_streichendes_fach}' streichst, verbessert sich der Schnitt auf {neuer_bester_schnitt:.3f} (Verbesserung: {beste_verbesserung:.2f})")
else:
    print("Das Streichen eines einzelnen Fachs führt zu keiner Verbesserung des Schnitts.")
```
